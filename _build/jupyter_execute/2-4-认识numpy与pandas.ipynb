### 11. 实现：连接数据帧

```python
df_1 = pd.DataFrame({
    'col1' : np.array([1, 2, 3]),
    'col2' : np.array(["A", "B", "C"])
})
df_2 = pd.DataFrame({
    'col1' : np.array([4, 5, 6]),
    'col2' : np.array(["D", "E", "F"])
})
```

```python
# 在纵向上连接
print(pd.concat([df_1, df_2]))
```

```python
# 在横向上连接
print(pd.concat([df_1, df_2], axis = 1))
```

### 12. 实现：数据帧的列操作

### 13. 实现：数据帧的行操作

```python
# 获取前 3 行
print(sample_df.head(n = 3))
```

```python
# 获取第 1 行
print(sample_df.query('index == 0'))
```

```python
# 通过多种条件获取数据
print(sample_df.query('col3 == "A"'))
```

```python
# 按 OR 条件获取数据
print(sample_df.query('col3 == "A" | col3 == "D"'))
```

```python
# 按 AND 条件获取数据
print(sample_df.query('col3 == "A" & col1 == 3'))
```

```python
# 同时指定行和列的条件
print(sample_df.query('col3 == "A"')[["col2", "col3"]])
```
